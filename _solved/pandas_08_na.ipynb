# Working with missing data

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A': [1, 2, np.nan],
                   'B': [4, np.nan, np.nan],
                   'C': [7, 8, 9]})
df
```

```python
df.loc[2, 'A']
```

```python
np.nan
```

```python
df['A'].mean()
```

```python
df['A'].mean(skipna=False)
```

```python
df['A'] + 3
```

```python
df['A'] == np.nan
```

```python
df['A'].isna() # .isnull()
```

```python
np.nan == np.nan
```

```python
df['A'].notna()
```

```python
df['A'].isna().sum()
```

```python
df.isna().sum()
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(subset=['A', 'C'])
```

```python
df.fillna()
```

```python
df.interpolate()
```

Summary:

* missing value indicator: `np.nan` (`NaN`)
* Reductions: skip by default
* Mathematical operations (eg `+`): propagate by default
* Specific functions:
    * `isna`, `notna`
    * `dropna`
    * `fillna`, `interpolate`
